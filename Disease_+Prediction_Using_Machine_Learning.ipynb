```python
train.drop('Unnamed: 133', axis=1, inplace=True)

# lets confirm that we have 133, and 134 columns respectively
len(train.columns), len(test.columns)
```

# split the data into test and train dataset

```python
y_train, y_test=train['prognosis'], test['prognosis']

X_train, X_test=train.drop('prognosis', axis=1), test.drop('prognosis', axis=1)
```

lets view that we have the correct shape for the y_train, and x_test

```python
y_train.shape, X_train.shape,
```

lets do the same for the test variables

```python
y_test.shape, X_test.shape
```

lets encode the labels

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train_encoded=encoder.fit_transform(y_train)
y_test_encoded=encoder.transform(y_test)
```

# model fitting

```python
# model 1 SVC

svc=SVC()
svc.fit( X_train,y_train_encoded)

preds=svc.predict(X_test)
accuracy_score(y_test_encoded, preds)
```
